```python
%matplotlib inline
#The line above is necesary to show Matplotlib's plots inside a Jupyter Notebook
import cv2
import numpy as np
from matplotlib import pyplot as plt
import time
start_time = time.time()

def remove_white_pixels(original_img, mask_img):
    
    # Copy 
    original_image_copy = original_img
    mask_image_copy = mask_img
    
    # Create X/Y boundaries for row and columns
    rows,cols,channel = mask_image_copy.shape
    
    # Get the background pixel value
    first_pixel_color_val = original_image_copy[0,0]
    
    # Ashton remove these two lines to implement into your code. 
    plt.imshow(cv2.cvtColor(original_image_copy, cv2.COLOR_BGR2RGB))
    plt.show()
    
    # Loop that iterates through each pixel, determines if white, then makes that pixel black. 
    for i in range(rows):
        for j in range(cols): 
            pixel_val = mask_image_copy[i,j]
            
            # Conditional to determine if pixel is white.
            if (np.array_equal(pixel_val, np.array([255,255,255]))):
                
            # Then eradicate white pixels. (make them black)
                 original_image_copy[i,j] = first_pixel_color_val
            
    print(first_pixel_color_val)        
    plt.imshow(cv2.cvtColor(original_image_copy, cv2.COLOR_BGR2RGB))
    plt.show()
    print("Time on this method: ", time.time() - start_time, " seconds")
    #------------------        
    #Random stuff
    #------------------
    # Potential way to read in files
    
    #  for name in [ "../c/pic%d.png" % i for i in [1, 2, 3, 4, 5, 6] ]:
        #img0 = cv.LoadImage(name, 1)
```
